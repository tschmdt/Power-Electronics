```python
#creating a sdynamic simulation of a FACTS device (power electronics) to show the behavior of the current for different firing angles alpha     
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Parameters
L = 0.1
R = 0.1
V = 1
f = 50
w = 2*np.pi*f
T = 1/f
t = np.linspace(0, 5*T, 1000)

# System of differential equations
def system(y, t, alpha):
    i, v = y
    di = (V*np.sin(w*t) - R*i - v)/L
    dv = (i - v)/R
    return [di, dv]

# Simulation
alpha = 60
y0 = [0, 0]
sol = odeint(system, y0, t, args=(alpha,))
i = sol[:,0]
v = sol[:,1]

# Plot
plt.plot(t, i, label='Current')
plt.plot(t, v, label='Voltage')
plt.legend()
plt.xlabel('Time [s]')
plt.ylabel('Current [A] / Voltage [V]')
plt.title('Current and Voltage of a FACTS device')
plt.grid()
plt.show()
```
