```python
import matplotlib.pyplot as plt
import numpy as np

# Daten für die typische Kennlinie
stromdichte = np.linspace(0, 1, 100)
spannung_typisch = 1.2 - 0.1 * stromdichte - 0.3 * np.log(stromdichte + 0.01)

# Daten für die geänderte Kennlinie bei starker Diffusionshemmung
spannung_diffusionshemmung = 1.2 - 0.1 * stromdichte - 0.3 * np.log(stromdichte + 0.01) - 0.5 * (stromdichte ** 2)

# Plotten der Kennlinien
plt.figure(figsize=(10, 6))
plt.plot(stromdichte, spannung_typisch, label='Typische Kennlinie', color='blue')
plt.plot(stromdichte, spannung_diffusionshemmung, label='Kennlinie bei Diffusionshemmung', color='red', linestyle='--')
plt.xlabel('Stromdichte (A/m²)')
plt.ylabel('Spannung (V)')
plt.title('Kennlinien einer Brennstoffzelle')
plt.legend()
plt.grid(True)
plt.show()

```
